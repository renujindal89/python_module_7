# sorted

```python
Dictionaries are key–value pairs, and you can’t use .sort() directly.
But you can use sorted() in three ways — on keys, values, or items.
```

```python
#Sort by Keys (Default)
d = {'a': 2, 'A': 3, 'c': 1}
print(sorted(d))  # ['a', 'b', 'c']  → only keys
```

```python
# Sorting a List of Dictionaries
```

```python
#Sort by a Single Key
people = [
    {"name": "Alice", "age": 25},
    {"name": "Bob", "age": 30},
    {"name": "Charlie", "age": 20}
]

sorted_people = sorted(people, key=lambda x: x["age"])
print(sorted_people)
```

```python
#Sort by a String Key (Alphabetically)
sorted_people = sorted(people, key=lambda x: x["name"])
print(sorted_people)
```

```python
#Sort in Reverse Order
sorted_people = sorted(people, key=lambda x: x["age"], reverse=True)
```
