```python
The enumerate() function in Python is a built-in function that adds a counter (index) 
to an iterable (like a list, tuple, or string)
```

```python
enumerate(iterable, start=0)
```

```python
fruits = ['apple', 'banana', 'cherry']

for index,value in enumerate(fruits):
    print(index, value)
```

```python
fruits = ['apple', 'banana', 'cherry']

for index, fruit in enumerate(fruits, start=100):
    print(index, fruit)
```

# sorting in Python for lists or tuples that contain integers or strings.

```python
#Sorting a List of Integers
numbers = [5, 2, 9, 1, 7]
sorted_numbers = sorted(numbers)  # returns a new sorted list
print(sorted_numbers)  # [1, 2, 5, 7, 9]

numbers.sort()  # sorts the list in place
print(numbers)  # [1, 2, 5, 7, 9]
```

```python
#Sorting a List of Strings
words = ["banana", "Apple", "cherry"]
print(sorted(words))  # ['apple', 'banana', 'cherry']
```

```python
#Case-insensitive sort:

print(sorted(words, key=str.lower))
```

```python
# Tuples themselves are immutable, so you cannot sort them in place
```

```python
t = (5, 2, 9, 1)
sorted_tuple = tuple(sorted(t))
print(sorted_tuple)  # (1, 2, 5, 9)
```

```python
#Sort by the Second Element
#Use the key argument to tell Python what to sort by.
#lambda used As a key for sorting: 
```

```python
data = [(3, 'apple'), (1, 'banana'), (2, 'cherry')]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
#Sort by the Second Element (Descending)
data = [(3, 'apple'), (1, 'banana'), (2, 'cherry')]
sorted_data = sorted(data, key=lambda x: x[1], reverse=True)
print(sorted_data)
```

```python
#Sort by Length of the Second Element
data = [(3, 'apple'), (1, 'kiwi'), (2, 'banana')]
sorted_data = sorted(data, key=lambda x: len(x[1]))
print(sorted_data)
```

```python
#Sort by Multiple Keys
#You can sort by more than one thing — for example, first by number, then by name:

data = [(2, 'banana'), (2, 'apple'), (1, 'cherry')]
sorted_data = sorted(data, key=lambda x: (x[0], x[1]))
print(sorted_data)
```

```python
#Sort List of Tuples Containing Mixed Types
#If you have tuples like (int, str) and want to sort safely even if elements differ in type:

data = [(3, 'apple'), ('1', 'banana'), (2, 'cherry')]
sorted_data = sorted(data, key=lambda x: (str(x[0]), x[1]))
print(sorted_data)
```

```python
# sorting lists of dictionaries or nested lists

#Sorting a List of Dictionaries
#Sort by a Single Key
people = [
    {"name": "Alice", "age": 25},
    {"name": "Bob", "age": 30},
    {"name": "Charlie", "age": 20}
]

sorted_people = sorted(people, key=lambda x: x["age"])  #{"name": "Alice", "age": 25},  25,30,20
print(sorted_people)
```

```python
#Sort by a String Key (Alphabetically)
sorted_people = sorted(people, key=lambda x: x["name"])
print(sorted_people)
```

```python
#Sort by Multiple Keys
#You can sort by one key primarily, then another key for tie-breaking.

people = [
    {"name": "Alice", "age": 25},
    {"name": "Alice", "age": 22},
    {"name": "Bob", "age": 30}
]

sorted_people = sorted(people, key=lambda x: (x["name"], x["age"]))
print(sorted_people)
```

```python
#Sort in Reverse Order
sorted_people = sorted(people, key=lambda x: x["age"], reverse=True)
```

```python
# Sorting a Nested List

#Sort by First Element
data = [[3, 'apple'], [1, 'banana'], [2, 'cherry']]
sorted_data = sorted(data, key=lambda x: x[0])
print(sorted_data)
```

```python
#Sort by Second Element
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
#Sort by Computed Value
#You can use any function in the key.
#Example — sort by length of the second element:

data = [[1, 'kiwi'], [2, 'banana'], [3, 'apple']]
sorted_data = sorted(data, key=lambda x: len(x[1]))
print(sorted_data)
```
