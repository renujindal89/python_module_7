# any

```python
Returns True if any element in an iterable is true.
Checks each element in the iterable from left to right.
Stops and returns True as soon as it finds a truthy element(true).
Returns False only if all elements are falsy.
```

```python
A value is truthy if it behaves like True
```

```python
A value is falsy if it behaves like False 
# anything which is empty or 0,[],{},"",set(),false,none
```

```python
values = [0, False, 3, "", []]
print(all(values))  # ✅ True (because 3 is truthy)
```

```python
values = [1, False, None, ""]
print(any(values))  # ❌ False (no truthy values)
```

```python
words = ["", "", "hello", ""]
print(any(words))  # True, because "hello" is a non-empty string (truthy)
```

# map()

```python
map() #Transforms every item  
🧠 Purpose: map() is great when you want to transform data.

Applies a function to each element of an iterable and returns a new iterator (you often convert it to a list).
```

```python
map(function, iterable) ,zip 
```

```python
# NOTE 
Returns an iterator, so you often wrap it with list() to see the result.

The function can be built-in, lambda, or user-defined.
```

```python
score = int(input("Enter your score: "))

if score >= 90 and score <= 100:
    grade = "A"
elif score >= 80 and score < 90:
    grade = "B"
elif score >= 70 and score < 80:
    grade = "C"
elif score >= 60 and score < 70:
    grade = "D"
elif score >= 0 and score < 60:
    grade = "F"
else:
    grade = "Invalid score"

print(f"Your grade is: {grade}")
```

```python
score=[1,54,87,97,56,34,98,89,90]
def get_grade(score):
    """Return a letter grade based on the numeric score."""
    if score >= 90 and score <= 100:
        return "A"
    elif score >= 80 and score < 90:
        return "B"
    elif score >= 70 and score < 80:
        return "C"
    elif score >= 60 and score < 70:
        return "D"
    elif score >= 0 and score < 60:
        return "F"
    else:
        return "Invalid score"
result=list(map(get_grade,score))   
print(result)
```

```python
functions like map() and zip() return iterators, not lists.
That means they don’t immediately produce all their results — they generate values on demand (lazy evaluation).

👉 So if you want to see, reuse, or index the results, you often need to wrap them in list().
```

```python
words = ["hello", "world", "python"]
uppercased = map(str.upper, words) #str.upper is a string method that converts a string to uppercase
print(list(uppercased))  # ['HELLO', 'WORLD', 'PYTHON']
```

```python
list(uppercased)
```

```python
names = ("alice", "bob", "carol")
capitalized = tuple(map(str.capitalize, names))
print(capitalized)  # ('Alice', 'Bob', 'Carol')
```

# filter()

```python
filter() #Keeps only items satisfied a condition #Subset of items
🧠 Purpose:

Filters (keeps) only elements of an iterable for which the function returns True
```

```python
filter(function, iterable)
```

```python
def ev(x):
    return x%2==0
num=[1,2,3,4,5,6,7,8]
a=list(filter(ev,num))
a
```

```python
def ev(x):
    return x%2==0
num=[1,2,3,4,5,6,7,8]
a=list(filter(ev,num))
a
```

```python
numbers = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  # [2, 4, 6]
```

```python
def is_positive(n):
    return n > 0

nums = [-3, -1, 0, 2, 4]
positives = filter(is_positive, nums)
print(list(positives))  # [2, 4]
#💡 filter() is great for selecting data.
```

```python
#Sort by the First Element
#first by the first element, then the second if the first is equal.
data = [(3, 'apple'), (3, 'banana'), (2, 'cherry')]
sorted_data = sorted(data)
print(sorted_data)
```

```python
#Sort by the Second Element
#Use the key argument to tell Python what to sort by.
#lambda used As a key for sorting: 
```

```python
data = [(3, 'apple'), (1, 'Banana'), (2, 'cherry')]
sorted_data = sorted(data, key=lambda x: x[1].lower())
print(sorted_data)
```

```python
#Sort by the Second Element (Descending)
data = [(3, 'apple'), (1, 'banana'), (2, 'cherry')]
sorted_data = sorted(data, key=lambda x: x[1], reverse=True)
print(sorted_data)
```

```python
#Sort by Length of the Second Element
data = [(3, 'apple'), (1, 'kiwi'), (2, 'banana')]
sorted_data = sorted(data, key=lambda x: len(x[1]))   #[5,4,6]
print(sorted_data)
```

```python
#Sort by Multiple Keys
#You can sort by more than one thing — for example, first by number, then by name:

data = [(2, 'banana'), (2, 'apple'), (1, 'cherry')]
sorted_data = sorted(data, key=lambda x: (x[0], x[1]))
print(sorted_data)
```

```python
# sorting lists of dictionaries or nested lists

#Sorting a List of Dictionaries
#Sort by a Single Key
people = [
    {"name": "Alice", "age": 25},
    {"name": "Bob", "age": 30},
    {"name": "Charlie", "age": 20}
]

sorted_people = sorted(people, key=lambda x: x["age"])  #{"name": "Alice", "age": 25},  25,30,20
print(sorted_people)
```
